```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('output.csv')
```

```python
df
```

```python
import seaborn as sns
for alg in df['algorithm'].unique():
    sns.lineplot(data=df[df.algorithm == alg], x='string_length', y='avg_time_ns', label=alg)
plt.legend()
plt.savefig('fig.png')
plt.show()
```
