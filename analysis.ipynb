```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('output.csv')
```

```python
df
```

```python
import os
if not os.path.isdir('images/'):
    os.makedirs('images/')
```

```python
import seaborn as sns

for alg in df['algorithm'].unique():
    sns.lineplot(data=df[df.algorithm == alg], x='string_length', y='avg_time_ns', label=alg)
plt.legend()
plt.savefig('images/fig1.png')
plt.show()
```

```python

for alg in df['algorithm'].unique():
    sns.lineplot(data=df[(df.algorithm == alg) & (df.pattern_length < 10)], x='string_length', y='avg_time_ns',
                 label=alg)
plt.legend()
plt.savefig('images/fig2.png')
plt.show()
```

```python
for alg in df['algorithm'].unique():
    sns.lineplot(data=df[(df.algorithm == alg) & (df.pattern_length == 50)], x='string_length', y='avg_time_ns',
                 label=alg)
plt.legend()
plt.savefig('images/fig3.png')
plt.show()
```
